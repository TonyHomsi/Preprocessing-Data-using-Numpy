    D) excludelist
    E) deletechars
    D) replace_space

```python
lending_co_TP = np.genfromtxt("Lending-Company-Total-Price.csv", 
                              delimiter = ',',
                              dtype = 'str',
                              skip_header = 1, 
                              skip_footer = 15,
                              usecols = (1,2,4),
                              comments = '#',
                              converters = None,
                              missing_values = None,
                              excludelist = None,
                              deletechars = " !#$%&'()*+,-./:;<=>?@[\\]^{|}~",
                              replace_space = '_',
                              autostrip = True
                             )
lending_co_TP
```

### 7. The first and last columns were the only ones with numeric data, so let's import only them.
